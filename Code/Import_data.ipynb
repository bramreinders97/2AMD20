Remove Irrelevant Townships (gemeentes)

```python
top_10_townships = ['GM0363','GM0599','GM0518','GM0344','GM0772','GM0855','GM0014','GM0034','GM0758','GM0268']
```

```python
moves_top_10_townships = moves_df.loc[(moves_df['RegioVanVestiging'].isin(top_10_townships)) | (moves_df['RegioVanVertrek'].isin(top_10_townships))]
#size (43500 x 5)
#correctness is checked
moves_top_10_townships
```

```python
house_prices_top_10_townships = house_prices_df.loc[house_prices_df['RegioS'].isin(top_10_townships)]
#size = (23 x 11)
#correctness is checked
#only includes Amsterdam, DenHaag, Rotterdam, Utrecht
```

```python
house_availability_top_10_townships = house_availability_df.loc[house_availability_df['RegioS'].isin(top_10_townships)]
#size = (180 x 10)
#correctness is checked
```

```python
population_top_10_townships = population_df.loc[population_df['RegioS'].isin(top_10_townships)]
#size = (150 x 24)
#correctness is checked
```

Rename GMxxxx to actual name of the township for the sake of clarity.

```python
#I include all townships because in the moves table the township that is being moved to/from is also included in the table
gemeentes = pd.read_excel("../Data/gemeentes.xlsx",header=None)
township_codes = gemeentes[0].to_list()
township_names = gemeentes[1].to_list()

def rename_township_values(df):
    return df.replace(township_codes, township_names)
```

```python
moves_top_10_townships_renamed = rename_township_values(moves_top_10_townships)
house_prices_top_10_renamed =  rename_township_values(house_prices_top_10_townships)
house_availability_top_10_renamed = rename_township_values(house_availability_top_10_townships)
population_top_10_renamed = rename_township_values(population_top_10_townships)

house_prices_top_10_renamed.RegioS.unique()
```
