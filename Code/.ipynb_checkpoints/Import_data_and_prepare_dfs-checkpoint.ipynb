## Import and clean moving data

```python
# Load data of movement between cities
moves_df = pd.read_pickle("../Prepared_DFs/moves.pkl")
conversions = pd.read_csv("../Data/Conversion_municipalities.csv")

# convert years to ints.
moves_df['Perioden'] = moves_df['Perioden'].map(lambda x: x.rstrip('00').rstrip("JJ"))
moves_df['Perioden'] = pd.to_datetime(moves_df['Perioden'], format='%Y').dt.year

# Conversions adjustment to strings
conversions = conversions.astype('string')
```

```python
# Get all null valued rows/indices
moves_null = moves_df[moves_df['TussenGemeentenVerhuisdePersonen_1'].isnull()]

# Convert all rows in moves_null. Updating the rows in moves_df. The next step will be to aggregate this.
for jndex, row2 in conversions.iterrows():
    for index, row in moves_df.iterrows():
        if row['RegioVanVestiging'] == row2['odCode']:
            moves_df.at[index, 'RegioVanVestiging'] = conversions.loc[jndex, 'newCode']
        if row['RegioVanVertrek'] == row2['odCode']:
            moves_df.at[index, 'RegioVanVertrek'] = conversions.loc[jndex, 'newCode']
```

```python
grouped = moves_df.groupby(['RegioVanVestiging', 'RegioVanVertrek', 'Perioden']).agg({'TussenGemeentenVerhuisdePersonen_1': 'sum'}).reset_index()
grouped
```

## Import and clean house_prices_new

```python
# Load housing prices
house_prices_new = pd.read_csv("../Data/house_prices_new.csv", sep = ';')

# Cut off the national and province level housing prices
house_prices_new = house_prices_new.loc[285:]
house_prices_new.reset_index(inplace=True)
house_prices_new.drop(columns = 'index',inplace=True)
```

```python
# Check if both datasets have the same amount of regions (they dont, moves has 209 less)
unique_house_prices = house_prices_new.RegioS.unique()
unique_moves = moves_df.RegioVanVestiging.unique()
difference = set(unique_house_prices) - set(unique_moves)

# Drop the difference from the house_prices_new dataset
difference_index = house_prices_new.loc[house_prices_new["RegioS"].isin(difference)].index
house_prices_new.drop(difference_index, inplace=True)
```

```python
# This is just to check whether we have dropped correctly. The datasets are now equal in size
unique_house_prices = house_prices_new.RegioS.unique()
unique_moves = moves_df.RegioVanVestiging.unique()
difference = set(unique_house_prices) - set(unique_moves)
print(len(unique_house_prices), len(unique_moves))
print(len(difference))
```

Rename GMxxxx to actual name of the township for the sake of clarity.

```python
#I include all townships because in the moves table the township that is being moved to/from is also included in the table
gemeentes = pd.read_excel("../Data/gemeentes.xlsx",header=None)
township_codes = gemeentes[0].to_list()
township_names = gemeentes[1].to_list()

def rename_township_values(df):
    return df.replace(township_codes, township_names)
```

```python
moves_top_10_townships_renamed = rename_township_values(moves_top_10_townships)
house_prices_top_10_renamed =  rename_township_values(house_prices_top_10_townships)
house_availability_top_10_renamed = rename_township_values(house_availability_top_10_townships)
population_top_10_renamed = rename_township_values(population_top_10_townships)
```
